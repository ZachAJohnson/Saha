```python
name='Xe 5 bar'
Z, A = 54, 131.29
N_ions = 5  # Number of ions, so N_ions + 1 = number of nuclear types
nn_invcc = 1.86e20  # 1/cc
T_K = 3e4 # Kelvin
T_AU = T_K*K_to_AU
rs = rs_from_n(nn_invcc*invcc_to_AU)
R = 10*rs

####
aa_kwargs = {'initialize':True, 'gradient_correction':None, 'μ_init' : 0.158, 'Zstar_init' : 'More', 'rmin':1e-2 ,'Npoints':300,
             'name':name, 'ignore_vxc':False, 'fixed_Zstar':False, 'iet_R_over_rs':R/rs*3, 'iet_N_bins':5000, 'use_full_ne_for_nf':False}


T_K_array = np.geomspace(1e3, 1e5, num=10)
aa_list = []
Zbar_list = []
for T_K in T_K_array:
    T_AU = T_K*K_to_AU
    
    aa = NPA(Z, A, T_AU, T_AU, rs, rs, **aa_kwargs)
    aa.solve_TF(verbose=False, picard_alpha=0.2, num_10folds_per_rs=0.5, tol=1e-4)
    
    aa_list.append(aa)
    Zbar_list.append(aa.Zstar)
    
```

```python
np.savetxt("Xenon_5bar_OFAA.dat", np.array([T_K_array, Zbar_list]).T, header='T[K]  Zbar  ', fmt='%10.6e', comments='')
```
